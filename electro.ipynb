```python
import basic_strategies as bs
import numpy as np 
import seaborn as sn
import pandas as pd

def round_robin(pl,it):
    results = []
    N=len(pl)
    for i in range(N):
        for j in range(N):
            results.append(bs.game(pl[i],pl[j],it))
            print( pl[i].k ,"-", pl[j].k , "  ", bs.game(pl[i],pl[j],it)[0], "-", bs.game(pl[i],pl[j],it)[1])
                
    return results
                
pl = [ bs.nice_guy, bs.bad_guy, bs.avg_guy, bs.mn_guy, bs.mb_guy, bs.tt_guy ]
results = round_robin(pl,100)

sum_scores = []
diff_scores = []
for score in results:
    sum_scores.append( score[0]+score[1] )
    diff_scores.append( score[0]-score[1] )

n=len(pl)
sum_scores = np.array(sum_scores).reshape(-n,n)
diff_scores = np.array(diff_scores).reshape(-n,n)

sum_matrix = pd.DataFrame(sum_scores/300, index = [i.k for i in pl],
                                          columns = [i.k for i in pl])

diff_scores = pd.DataFrame(diff_scores/300, index = [i.k for i in pl],
                                          columns = [i.k for i in pl])
```

```python
sn.heatmap(sum_matrix, annot=True, cmap="Greens")
```

```python
sn.heatmap(diff_scores, annot=True, cmap="Blues")
```
